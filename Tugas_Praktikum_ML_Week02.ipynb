# **1. Memisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.**

Dalam pemilihan variabel yang dapat digunakan dalam kasus ini bisa menggunakan teknik Recursive Feature Elimination (RFE).  Teknik RFE dilakukan dengan melakukan analisis eksplorasi data dan memeriksa korelasi antara setiap fitur/variabel dan diagnosis kanker. Fitur-fitur dengan korelasi tinggi dengan diagnosis kanker kemungkinan besar akan menjadi fitur/variabel yang berguna dalam memprediksi diagnosis kanker.

Maka, didapatkan variabel yang berguna dalam kasus ini, diantaranya:
1. **area_worst**
2. **concave points_worst**
3. **texture_worst**
4. **radius_worst**
5. **perimeter_worst**

# **4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.**

Load data

```python
dpath = '/content/drive/MyDrive/Machine Learning/wbc_dataset_selected.csv'
df = pd.read_csv(dpath)
df.head()
```

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
df.head()
```

```python
std = StandardScaler()
df['radius_worst'] = std.fit_transform(df[['radius_worst']])
df['texture_worst'] = std.fit_transform(df[['texture_worst']])
df['perimeter_worst'] = std.fit_transform(df[['perimeter_worst']])
df['area_worst'] = std.fit_transform(df[['area_worst']])
df['concave points_worst'] = std.fit_transform(df[['concave points_worst']])
```

```python
df = df[['diagnosis', 'area_worst', 'concave points_worst', 'texture_worst', 'radius_worst', 'perimeter_worst']]
df.head()
```

Split data

```python
from sklearn.model_selection import train_test_split

df2 = pd.read_csv('/content/drive/MyDrive/Machine Learning/wbc_dataset_selected.csv')
X = df2.drop('diagnosis', axis=1)
y = df2['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)  #test_size=0.2 untuk menentukan bahwa data uji (test) sebesar 20% dan data latih (train) sebesar 80%
```
